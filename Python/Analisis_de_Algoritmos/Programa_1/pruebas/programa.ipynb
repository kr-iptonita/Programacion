```python
import tkinter as tk
import colorsys

# Definir la matriz
arr = [[33, 33, 32, 32], [33, 31, -1, 32], [34, 31, 31, 35], [34, 34, 35, 35]]
n = len(arr)



# Función para convertir un número en un color hexadecimal
def number_to_color(number):
    if number == -1:
        return "#000000"  # negro para -1
    else:
        hue = number / 100.0  # utilizar el número para determinar el tono
        rgb = colorsys.hsv_to_rgb(hue, 1, 1)  # convertir de HSV a RGB
        r, g, b = [int(255 * x) for x in rgb]  # convertir de flotante a entero
        return f"#{r:02x}{g:02x}{b:02x}"  # convertir a hexadecimal y devolver

# Crear la ventana y el lienzo
root = tk.Tk()
canvas = tk.Canvas(root, width=400, height=400)
canvas.pack()

# Dibujar la cuadrícula
for i in range(n):
    for j in range(n):
        x1 = j * 100
        y1 = i * 100
        x2 = x1 + 100
        y2 = y1 + 100
        color = number_to_color(arr[i][j])
        canvas.create_rectangle(x1, y1, x2, y2, fill=color)

# Iniciar la aplicación
root.mainloop()
```
